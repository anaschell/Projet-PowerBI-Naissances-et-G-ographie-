```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("merged_file.csv", delimiter=';')
```

```python
df.describe()
```

```python
sex_mapping = {1: 'homme', 2: 'femme'}
df['sexe'] = df['sexe'].replace(sex_mapping)
df.head()
```

```python
non_string_indices = df[~df['sexe'].apply(lambda x: isinstance(x, str))].index
df = df.drop(non_string_indices)
df.head()
```

```python
non_string_indices = df[~df['preusuel'].apply(lambda x: isinstance(x, str))].index
df = df.drop(non_string_indices)
df.head()
```

```python
non_integer_indices = df[~df['annais'].astype(str).str.isdigit()].index
df = df.drop(non_integer_indices)
df['annais'] = df['annais'].astype(int)
df.head()
```

```python
non_integer_indices = df[~df['dpt'].astype(str).str.isdigit()].index
df = df.drop(non_integer_indices)
df['dpt'] = df['dpt'].astype(int)
df.head()
```

```python
non_integer_indices = df[~df['nombre'].astype(str).str.isdigit()].index
df = df.drop(non_integer_indices)
df['nombre'] = df['nombre'].astype(int)
df.head()
```

```python
duplicate_rows = df[df.duplicated()]

# Display duplicate rows
print(f"Duplicate Rows: {duplicate_rows}")
```

```python
df.to_csv('updated_merged_file.csv', index=False)
```
